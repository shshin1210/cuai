```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
import pandas as pd
import urllib.request

URL = 'https://comic.naver.com/webtoon/finish'
html = requests.get(URL).text # html 문서 전체를 긁어서 출력해줌, .text는 태그 제외하고 text만 출력되게 함
soup = BeautifulSoup(html, 'html.parser')
num=0
```

```python
store_list = []
```

```python
from webdriver_manager.chrome import ChromeDriverManager

driver=webdriver.Chrome(ChromeDriverManager().install())

driver.implicitly_wait(10)
driver.get(URL)
```

```python
page = driver.find_elements_by_xpath('//a[contains(@onclick,"lst.title")]')
```

```python
store = soup.find_all('dt') # 두 번째 h2태그에 있음

for i in range(len(page)):
    
    if '스토어' in str(store[i]):
        store_list.append('2')

    else:
        store_list.append('1')
```
